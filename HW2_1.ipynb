```python
import pandas as pd
url_t = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
                   header=0, thousands=".", decimal=',')
df = url_t[12]
print(df)
```

```python
# Визначте кількість рядків та стовпців у датафреймі
print(df.shape)
```

```python
# Замініть у таблиці значення "—" на значення NaN\
new_df = df.replace('—', 'NaN')
print(new_df)
```

```python
# Визначте типи всіх стовпців за допомогою
new_df.dtypes
```

```python
# Замініть типи нечислових колонок на числові.
from numpy import float64
new_df.iloc[:, 1:] = new_df.iloc[:, 1:].astype(float)
print(new_df)
new_df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці
print(new_df.isnull().sum())
print(new_df.isnull().sum()/len(new_df)*100)
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
print(new_df.drop(new_df.index[-1]))
type(new_df)
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців

new_df_1 = new_df.fillna(new_df.mean().round(3))
new_df_1
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
av = new_df_1['2019'].mean()
print(av)
av_n = new_df[new_df['2019'] > av]
# regions_with_higher_birth_rate = df[(df['Рік'] == 2019) & (df['Рівень_народжуваності'] > average_birth_rate_ukraine_2019)]['Регіон']

print(av_n[['Регіон', '2019']])
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
new_df_max = new_df['2014'].idxmax()
reg_name = new_df.loc[new_df_max, 'Регіон']
print(f'{reg_name} область, мала найвищу народжуваність у 2014 році')


```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
import matplotlib.pyplot as plt
birtt_2019 = new_df["2019"]
regions = new_df["Регіон"]

plt.figure(figsize=(12, 6))
bars = plt.bar(regions, birtt_2019, color="b", edgecolor="k")
plt.xlabel("Регіони")
plt.ylabel("Рівень народжуваності у 2019 році")
plt.title("Рівень народжуваності по регіонах у 2019 році")
plt.xticks(rotation=90)
plt.tight_layout()
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             f'{birtt_2019[i]:.1f}', ha='center', va='bottom')
plt.show()
```
