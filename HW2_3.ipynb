```python
import pandas as pd
```

```python
df = pd.read_csv('https://drive.google.com/uc?id=16j0SIsf2TRRcveNS5ZaIfEG55rXdEZwL')
```

```python
print(f'Про скільки книг зберігає дані датасет? \nПро {df.shape[0]} книг')
```

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
df.head()
```

```python
pass_df = df.isna().sum()
print(f"Чи є в якихось змінних пропуски? \n{pass_df.any()}")
```

```python
g_u = df['genre'].unique()
print(f"Які є унікальні жанри? \n{g_u}")
```

```python
import matplotlib.pyplot as plt
df["price"].plot(kind='hist', bins=20, edgecolor='black')
plt.title("Розподіл цін")
plt.xlabel("Ціна")
plt.ylabel("Частота")

plt.show()
```

```python
max_price = df["price"].max()
min_price = df["price"].min()
mean_price = df["price"].mean()
mediaan_price = df["price"].median()


print(f"Максимальна ціна? \n{max_price}")
print(f"Мінімальна ціна? \n{min_price}")
print(f"Середня ціна? \n{mean_price}")
print(f"Медіанна ціна? \n{mediaan_price}")
```

```python
max_user_rating = df["user_rating"].max()

print(f"Який рейтинг у датасеті найвищий? \n{max_user_rating}")
```

```python
max_user_rating_books = df[ df["user_rating"] == max_user_rating]

print(f"Скільки книг мають такий рейтинг? \n{max_user_rating_books.shape[0]}")
```

```python
max_reviews_index = df["reviews"].idxmax()
book_with_most_reviews = df.loc[max_reviews_index]

print(f'Яка книга має найбільше відгуків? \n"{book_with_most_reviews[0]}"')
```

```python
top_books_2015 = df[df["year"] == 2015]
max_price_index = top_books_2015['price'].idxmax()
most_expensive_book = top_books_2015.loc[max_price_index]
print(f' З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожч?  \n{most_expensive_book[0]}\n(Вартість: {most_expensive_book["price"]}$)')
```

```python
fiction_book_in_2010 = df[(df["genre"] == "Fiction") & (df["year"] == 2010)]

print(f"Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році? \n{fiction_book_in_2010.shape[0]} книг")
```

```python
user_rating_book_in_years = df[(df["user_rating"] == "4.9") & (df["year"].isin([2010, 2011]))]
print(f"Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках? \n{user_rating_book_in_years.shape[0]} книг")
```

```python
table_books_2015 = df[(df["year"] == 2015) & (df["price"] < 8)].sort_values(by="price")
table_books_2015
```

```python
table_books_2015.tail(1)
```

```python
table_books_genre = df[['genre', 'price']].groupby('genre').agg(["min", "max"])

max_price_fiction = table_books_genre.loc['Fiction', ('price', 'max')]
min_price_fiction = table_books_genre.loc['Fiction', ('price', 'min')]
max_price_non_fiction = table_books_genre.loc['Non Fiction', ('price', 'max')]
min_price_non_fiction = table_books_genre.loc['Non Fiction', ('price', 'min')]

print(f"Максимальна ціна для жанру Fiction? \n{max_price_fiction}")
print(f"Мінімальна ціна для жанру Fiction? \n{min_price_fiction}")
print(f"Максимальна ціна для жанру Non Fiction? \n{max_price_non_fiction}")
print(f"Мінімальна ціна для жанру Non Fiction? \n{min_price_non_fiction}")
```

```python
author_book_counts = df.groupby('author')['name'].count().reset_index()

print(author_book_counts)

print(f"Якої розмірності вийшла таблиця? \n{author_book_counts.shape}")
print(f"Який автор має найбільше книг? \n{author_book_counts['author'].max()}")
print(f"Скільки книг цього автора? \n{author_book_counts['name'].max()} books")
```

```python
author_avg_rating = df.groupby('author')['user_rating'].mean().reset_index()

min_mean_author = author_avg_rating.loc[author_avg_rating['user_rating'].idxmin(), 'author']

print(f"У якого автора середній рейтинг мінімальний? \n{min_mean_author}")
print(f"Який у цього автора середній рейтинг? \n{author_avg_rating['user_rating'].min()}")
```

```python
author_book_counts = author_book_counts.rename(columns={'name': 'Count_of_books'})
combine_table = pd.concat([author_avg_rating, author_book_counts['Count_of_books']], axis=1)

print(combine_table.head())
```

```python
combine_table.iloc[0]
```

```python
sorted_combine_table = combine_table.sort_values(by=['Count_of_books', 'user_rating'], ascending=[True, True])
sorted_combine_table.iloc[0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(8, 50))
sns.barplot(x='Count_of_books', y='author', data=sorted_combine_table, orient='h')

plt.title('Кількість книг за авторами')
plt.xlabel('Кількість книг')
plt.ylabel('Автори')

plt.tight_layout()
plt.show()
```
